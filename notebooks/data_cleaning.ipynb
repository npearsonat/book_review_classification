```python
import pandas as pd
import numpy as np
```

```python
df_book = pd.read_csv('data/books_data.csv')
df_rate = pd.read_csv('data/Books_rating.csv')
```

```python
df_rate.head(5)
```

```python
# Get rid of reviews with less than 10 total review/helpfulness reviws
df_rate = df_rate[df_rate["review/helpfulness"].str.split("/").str[1].astype(int) >= 10]

df_rate["review/helpfulness"] = (
    df_rate["review/helpfulness"]
    .str.split("/")
    .apply(lambda x: int(x[0]) / int(x[1]))
)
```

The list of ratings is over 3 million rows, so I'll take a random assortment of 100k

```python
# median (0.8) as cutoff for balanced classes 
df_rate['is_helpful'] = (df_rate['review/helpfulness'] > 0.8).astype(int)
df = df_rate.groupby('is_helpful', group_keys=False).apply(lambda x: x.sample(frac=0.025, random_state=42))
```

```python
df
```

```python
# adding df_book information as a right join
df_merged = df_book.merge(df, on="Title", how="right")
```

```python
# Timestamp feature engineering. Converting unix time to datetime and creating separate categories
df["datetime"] = pd.to_datetime(df["review/time"], unit="s")

df["year"] = df["datetime"].dt.year
df["month"] = df["datetime"].dt.month
df["day"] = df["datetime"].dt.day
df["day_of_week"] = df["datetime"].dt.dayofweek   # Monday=0, Sunday=6
# df["hour"] = df["datetime"].dt.hour # hour has no variation, all at 0:00
df["is_weekend"] = df["day_of_week"].isin([5,6]).astype(int)
df["quarter"] = df["datetime"].dt.quarter
```

```python
# drop price, user_id_ and profie_Name column, then get rid of the few remaining nulls
df = df.drop(columns=["Price", "User_id", "profileName","Id","Title","review/score"])
df = df.dropna()
```

```python
df_nulls = df.isnull().sum().reset_index()
df_nulls.columns = ["column", "null_count"]
print(df_nulls)
```

```python
df.to_csv("data_cleaned.csv", index=False)
```
